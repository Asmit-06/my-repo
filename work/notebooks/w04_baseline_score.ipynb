## 1. My rule and its reason codes

*Write the rule in plain words first. Then the reason codes it can output.*

## My Rule

I will prioritize pages for content refresh based on two signals: days since the last update and impressions over the last 90 days. Pages that have not been updated for at least 180 days and still receive high impressions are more valuable to refresh because they have greater visibility and a higher potential impact.

## Reason Codes

- STALE_HIGH_VISIBILITY – The page is stale and receives high impressions.
- STALE_LOW_VISIBILITY – The page is stale but has low impressions.
- RECENT_HIGH_VISIBILITY – The page is recently updated but still receives high impressions.
- LOW_PRIORITY – The page is neither stale nor highly visible.

## 2. Build the ranked queue (writes the CSV)

*Code the score, rank everything, write work/outputs/baseline_action_score.csv.*

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd
import os

# Example if df is already loaded
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")
df["baseline_score"] = (
    (df["days_since_last_update"] >= 180).astype(int) * 2 +
    (df["impressions_90d"] >= 500).astype(int)
)

def reason(row):
    if row["days_since_last_update"] >= 180 and row["impressions_90d"] >= 500:
        return "STALE_HIGH_VISIBILITY"
    elif row["days_since_last_update"] >= 180:
        return "STALE_LOW_VISIBILITY"
    elif row["impressions_90d"] >= 500:
        return "RECENT_HIGH_VISIBILITY"
    else:
        return "LOW_PRIORITY"

df["reason_code"] = df.apply(reason, axis=1)

df["action"] = df["reason_code"].apply(
    lambda x: "Refresh Content"
    if x == "STALE_HIGH_VISIBILITY"
    else "Monitor"
)

ranked = df.sort_values(
    "baseline_score",
    ascending=False
)

os.makedirs("work/outputs", exist_ok=True)

ranked.to_csv(
    "work/outputs/baseline_action_score.csv",
    index=False
)

ranked.head()
```
